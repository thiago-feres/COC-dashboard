```python
import sqlite3
import pandas as pd

# Conecta ao banco
conn = sqlite3.connect("../data/clash_history.db")

# Query "Poderosa": Une as duas tabelas usando a TAG como ponte
query = """
SELECT 
    h.*, 
    p.townHallLevel,
    p.bestTrophies, 
    p.warStars, 
    p.rei, 
    p.rainha,
    p.principe_servo,
    p.guardiao,
    p.campea_real

FROM historico_membros h
LEFT JOIN player_details p ON h.tag = p.tag
ORDER BY h.coletado_em DESC
"""

# Lê os dados
df_completo = pd.read_sql(query, conn)

conn.close()

# Configuração para o Pandas não esconder colunas se houver muitas
pd.set_option('display.max_columns', None)

# Exibe os últimos 50 registros (com todas as colunas lado a lado)
df_completo.head(50)
```
